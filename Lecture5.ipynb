```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 5 random blue points
blue_points = np.random.rand(5, 2)

# Generate 4 random green points
green_points = np.random.rand(4, 2)

# Plot blue points
plt.scatter(blue_points[:, 0], blue_points[:, 1], color='blue')

# Plot green points
plt.scatter(green_points[:, 0], green_points[:, 1], color='green')

# Plot translucent blue disks around blue points
max_radius = 0
for blue_point in blue_points:
    max_dist = 0
    for green_point in green_points:
        distance = np.linalg.norm(blue_point - green_point)
        if distance > max_dist:
            max_dist = distance
    if max_dist > max_radius:
        max_radius = max_dist
    circle = plt.Circle((blue_point[0], blue_point[1]), max_dist, color='blue', alpha=0.1)
    plt.gca().add_patch(circle)

plt.axis('off')
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ripser import ripser
from ripser import Rips

# Function to generate noisy points on a unit circle
def generate_points(num_points, noise=0.1):
    angles = np.linspace(0, 2*np.pi, num_points)
    points = np.array([np.cos(angles), np.sin(angles)]).T
    noise_points = noise * np.random.randn(num_points, 2)
    return points + noise_points

# Generate noisy points on a unit circle
points = generate_points(100)

# Compute persistence homology using Ripser
rips = Rips()
diagrams = rips.fit_transform(points)

# Plot point cloud and persistence diagram side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot point cloud
axs[0].scatter(points[:, 0], points[:, 1], color='blue', alpha=0.5)
axs[0].set_aspect('equal', adjustable='box')

# Plot persistence diagram
rips.plot(diagrams, ax=axs[1])

plt.tight_layout()
plt.show()
```
