```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate clustered data
num_points = 24
num_clusters = 3
cluster_std_dev = 1.0

# Generate data points grouped in clusters
X, _ = make_blobs(n_samples=num_points, centers=num_clusters, cluster_std=cluster_std_dev)

# Create linkage
Z = linkage(X, method='single', metric='euclidean')

# Plot points and dendrogram
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].scatter(X[:, 0], X[:, 1], c='b', label='Points')

dendrogram(Z, ax=axs[1])

plt.tight_layout()
plt.show()
```

In the following diagram we observe that data of very different shapes can have the same dendrogram. In this case, the collinear points all cluster at scale 1, as do the honeycomb points, resulting in the left and right portions of the dendrogram looking identical.

```python
line_points_x = np.array([-3, -2, -1, 0, 1, 2, 3])
line_points_y = np.full_like(line_points_x, 3)

hexagon_points_x = np.cos(np.linspace(0, 2*np.pi, 7))
hexagon_points_y = np.sin(np.linspace(0, 2*np.pi, 7))

all_points_x = np.concatenate([line_points_x, hexagon_points_x, np.array([0])])
all_points_y = np.concatenate([line_points_y, hexagon_points_y, np.array([0])])

points = np.column_stack((all_points_x, all_points_y))

Z = linkage(points, method='single', metric='euclidean')

# Plot points and dendrogram
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].scatter(points[:, 0], points[:, 1], c='b')

dendrogram(Z, ax=axs[1])

plt.tight_layout()
plt.show()
```
